On peut maintenant faire tourner notre programme. Pour cela, il vous suffira d'écrire le nom de l'image que vous voulez modifier à la ligne 2 (en prenant une des images proposées dans le dossier `images`) et de modifier les couleurs à la ligne 11.

*Attention !* Toutes vos images modifiées auront le même nom si vous ne modifiez pas aussi la ligne 13)

```python
from PIL import Image
img = Image.open('images/boole.jpg')
display(img)

largeur, hauteur = img.size
print('Nombre de pixels en largeur :', largeur)
print('Nombre de pixels en hauteur :', hauteur)

for x in range(largeur):
    for y in range(hauteur):
        R, V, B = img.getpixel((x, y))
        img.putpixel((x, y),(R, V, 0))

img.save("nouvelle_image.jpg")
display(img)
```
