1. Achromatopsie

Pour récupérer la composante en rouge, vert et bleu d'un pixel.

```python
from PIL import Image
img = Image.open('daltonisme/couleurs.bmp')
display(img)
x = 20
y = 20
R, V, B = img.getpixel((x, y))
print(R, V, B)
```

Pour transformer un pixel en gris.

```python
from PIL import Image
img = Image.open('daltonisme/couleurs.bmp')
display(img)
x = 20
y = 20
R, V, B = img.getpixel((x, y))
print(R, V, B)
Gris = 0.3*R + 0.59*V + 0.11*B
print("Valeur de gris :", Gris)
Gris = int(Gris)
print("Valeur entière de gris :", Gris)
img.putpixel((x, y),(Gris, Gris, Gris))
display(img)
img.save("daltonisme/un_pixel_en_gris.png")
```
